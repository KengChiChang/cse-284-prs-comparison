```python
import os
```

```python
CT_files = os.listdir("/content/CT/")
BayesR_files = os.listdir("/content/BayesR/")
BASIL_files = os.listdir("/content/BASIL/")
len(CT_files), len(BayesR_files), len(BASIL_files)
```

```python
import pandas as pd

def ParseFileList(file_list):
    col_names = ['combined', 'type', 'superpop', 'power', 'her', 'num-causals', 'pheno']
    df = pd.DataFrame({"filename": file_list})
    df[col_names] = df['filename'].str.split('_',expand=True)
    df[['pheno', 'ext']] = df['pheno'].str.split('.',expand=True)
    return df
```

```python
CT_files_df = ParseFileList(CT_files)
BayesR_files_df = ParseFileList(BayesR_files)
BASIL_files_df = ParseFileList(BASIL_files)
CT_files_df.head()
```

```python
(CT_files_df.query('type=="predict"')
    .query('type=="predict"')
    .groupby(['superpop','power', 'her'])
    .size().reset_index(name='count')
    .sort_values(['superpop','power', 'her'])
)
```

```python
(BayesR_files_df
    .query('type=="predict"')
    .groupby(['superpop','power', 'her'])
    .size().reset_index(name='count')
    .sort_values(['superpop','power', 'her'])
)
```

```python
(BASIL_files_df.query('type=="predict"')
    .query('type=="predict"')
    .groupby(['superpop','power', 'her'])
    .size().reset_index(name='count')
    .sort_values(['superpop','power', 'her'])
)
```
