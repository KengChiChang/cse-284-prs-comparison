```python
sh = """


BED=EUR_train
PHENO=power=-1_her=0.5_num-causals=1000.phen
OUT=BayesR_EUR_train_power=-1_her=0.5_num-causals=1000

./gctb_2.05beta_Linux/gctb \
     --bayes R \
     --bfile $BED \
     --pheno $PHENO \
     --original-model \
     --chain-length 1000 \
     --burn-in 200 \
     --out-freq 10 \
     --out $OUT

"""
with open('script.sh', 'w') as file:
  file.write(sh)

! bash script.sh
```

```python
sh = """

TRAIN=EUR_train
MODEL=BayesR_EUR_train_power=-1_her=0.5_num-causals=1000.snpRes
OUT=BayesR_EUR_train_score_power=-1_her=0.5_num-causals=1000

./plink \
    --bfile $TRAIN \
    --score $MODEL 2 5 8 header sum center \
    --allow-no-sex \
    --maf 0.05 \
    --out $OUT

"""
with open('script.sh', 'w') as file:
  file.write(sh)

! bash script.sh
```

```python
import pandas as pd
import scipy.stats
prs = pd.read_csv("BayesR_EUR_train_score_power=-1_her=0.5_num-causals=1000.profile", delim_whitespace=True)
# phen = pd.read_csv("power=-1_her=0.5_num-causals=1000.phen",
#                    delim_whitespace=True)
phen = pd.read_csv("power=-1_her=0.5_num-causals=1000.phen",
                   delim_whitespace=True, names=["FID","IID"] + ["PHENO" + str(i) for i in range(1, 101)])
d = pd.merge(prs, phen, on=["IID"])
d
```

```python
print("EUR - test on EUR")
print("R2=%s"%(scipy.stats.pearsonr(d["PHENO1"],d["SCORESUM"])[0]**2))
```
