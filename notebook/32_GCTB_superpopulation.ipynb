```python
import pandas as pd
import scipy.stats
import warnings
warnings.filterwarnings('ignore')
```

```python
LOG_ID           = '2024-3-10-1'
POP_LIST         = ['EUR', 'AFR', 'EAS', 'SAS', 'AMR']
POWER_LIST       = [-0.25, -1]
HER_LIST         = [0.1, 0.5, 0.9]
NUM_CAUSALS_LIST = [1, 10, 100, 512]
PHENO_RANGE      = '1-20'

CHR19             = '/content/chr19'
SPLIT_DIR         = '/content/split_ids/by_superpopulation'
GENO_DIR          = '/content/chr19_by_superpopulation'
PHEN_DIR          = '/content/chr19_ldl_pheno'
MODEL_DIR         = '/content/chr19_ldl_pheno_result/models'
SCORE_DIR         = '/content/chr19_ldl_pheno_result/scores'
RESULT_DIR        = '/content/chr19_ldl_pheno_result'
GDRIVE_RESULT_DIR = '/content/drive/MyDrive/CSE-284-Final-Project/data/BayesR/chr19_ldl_pheno'

for POP in POP_LIST:
    print(f"-------------- Start: {POP} --------------")
    # convert superpopulation genotype data to .bed and do train test split
    for SPLIT in ['train', 'val', 'test']:
        sh = f'''
        ./plink2 \
            --make-bed \
            --bfile {CHR19} \
            --keep {SPLIT_DIR}/{POP}_{SPLIT}_IID_only.txt \
            --out {GENO_DIR}/{POP}_{SPLIT}
        '''
        with open('split_pop_bed.sh', 'w') as file:
            file.write(sh)
        ! bash split_pop_bed.sh >> GCTB_{LOG_ID}.log

    # train BayesR model and save batched result
    for POWER in POWER_LIST:
        for HER in HER_LIST:
            for NUM_CAUSALS in NUM_CAUSALS_LIST:

                PHENO_INDEX_START        = int(PHENO_RANGE.split('-')[0])
                PHENO_INDEX_END          = int(PHENO_RANGE.split('-')[1])
                combined_result_file     = f'{RESULT_DIR}/combined_result_{POP}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS}_pheno={PHENO_INDEX_START}-{PHENO_INDEX_END}.csv'
                combined_prediction_file = f'{RESULT_DIR}/combined_predict_{POP}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS}_pheno={PHENO_INDEX_START}-{PHENO_INDEX_END}.csv'
                combined_result_data     = []
                combined_prediction_data = []

                # get models, one for each phenotype
                for PHENO_INDEX in range(PHENO_INDEX_START, PHENO_INDEX_END + 1):
                    print(f"---- Train: pop={POP}, power={POWER}, her={HER}, num-causals={NUM_CAUSALS}, pheno={PHENO_INDEX}")

                    sh = f'''
                    ./gctb_2.05beta_Linux/gctb \
                        --bayes R \
                        --bfile {GENO_DIR}/{POP}_train \
                        --pheno {PHEN_DIR}/power={POWER}_her={HER}_num-causals={NUM_CAUSALS}.pheno \
                        --original-model \
                        --seed 123 \
                        --pi 0.95,0.02,0.02,0.01 \
                        --gamma 0.0,0.01,0.1,1 \
                        --chain-length 500 \
                        --burn-in 200 \
                        --out-freq 50 \
                        --mpheno {PHENO_INDEX} \
                        --out {MODEL_DIR}/BayesR_{POP}_train_PHENO{PHENO_INDEX}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS}
                    '''
                    with open('train_BayesR_model.sh', 'w') as file:
                        file.write(sh)
                    ! bash train_BayesR_model.sh >> GCTB_{LOG_ID}.log

                    # score train and test sets by trained model
                    for SPLIT in ['train', 'test']:
                        print(f"---- Score: split={SPLIT}, pop={POP}, power={POWER}, her={HER}, num-causals={NUM_CAUSALS}, pheno={PHENO_INDEX}")
                        sh = f'''
                        ./plink \
                            --bfile {GENO_DIR}/{POP}_{SPLIT} \
                            --score {MODEL_DIR}/BayesR_{POP}_train_PHENO{PHENO_INDEX}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS}.snpRes 2 5 8 header sum center \
                            --out {SCORE_DIR}/BayesR_{POP}_{SPLIT}_PHENO{PHENO_INDEX}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS} \
                            2> /dev/null
                        '''
                        with open('score_train_test.sh', 'w') as file:
                            file.write(sh)
                        ! bash score_train_test.sh >> GCTB_{LOG_ID}.log

                    # get combined prediction and result data for a particular phenotype
                    train_scores = []
                    test_scores = []
                    train_score = 0
                    test_score = 0

                    pheno_file = f'{PHEN_DIR}/power={POWER}_her={HER}_num-causals={NUM_CAUSALS}.pheno'
                    phen = pd.read_csv(pheno_file, delim_whitespace=True,
                                       usecols=[0, 1, PHENO_INDEX + 1], names=["FID","IID","phen"])

                    # calculate train score for a particular phenotype
                    train_score_file = f'{SCORE_DIR}/BayesR_{POP}_train_PHENO{PHENO_INDEX}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS}.profile'
                    train_prs = pd.read_csv(train_score_file, delim_whitespace=True)
                    train_score_df = pd.merge(train_prs, phen, on=["FID", "IID"])
                    train_score = scipy.stats.pearsonr(train_score_df["phen"], train_score_df["SCORESUM"])[0]**2
                    print(f"-- Train score {train_score}: pop={POP}, power={POWER}, her={HER}, num-causals={NUM_CAUSALS}, pheno={PHENO_INDEX}")

                    # calculate test score for a particular phenotype
                    test_score_file = f'{SCORE_DIR}/BayesR_{POP}_test_PHENO{PHENO_INDEX}_power={POWER}_her={HER}_num-causals={NUM_CAUSALS}.profile'
                    test_prs = pd.read_csv(test_score_file, delim_whitespace=True)
                    test_score_df = pd.merge(test_prs, phen, on=["FID", "IID"])
                    test_score = scipy.stats.pearsonr(test_score_df["phen"], test_score_df["SCORESUM"])[0]**2
                    print(f"-- Test score {test_score}: pop={POP}, power={POWER}, her={HER}, num-causals={NUM_CAUSALS}, pheno={PHENO_INDEX}")

                    predict_df = (test_score_df[['IID', 'SCORESUM', 'phen']]
                                  .rename(columns={'IID': 'IID',
                                                   'SCORESUM': 'predict',
                                                   'phen': 'actual'})
                                  .assign(pheno_num=PHENO_INDEX))
                    combined_prediction_data.append(predict_df)

                    result_dict = {
                        "population/superpopulation": POP,
                        "power": POWER,
                        "her": HER,
                        "num_causals": NUM_CAUSALS,
                        "pheno_num": PHENO_INDEX,
                        "train": train_score,
                        "test": test_score
                    }
                    combined_result_data.append(result_dict)

                combined_prediction_df = pd.concat(combined_prediction_data, ignore_index=True)
                combined_result_df = pd.DataFrame(combined_result_data)

                # Write combined prediction data to file
                combined_prediction_df.to_csv(combined_prediction_file, index=False)
                print("-- Combined prediction data has been written to:", combined_prediction_file)

                # Write combined result data to file
                combined_result_df.to_csv(combined_result_file, index=False)
                print("-- Combined result data has been written to:", combined_result_file)

                ! cp {combined_result_file} {GDRIVE_RESULT_DIR}
                ! cp {combined_prediction_file} {GDRIVE_RESULT_DIR}

            ! rm -rf {MODEL_DIR}/BayesR*

    ! rm -rf {GENO_DIR}/{POP}*
```

```python
sh = """

SUPERPOP=EUR
GENO_DIR=/content/chr19_by_superpopulation
PHEN_DIR=/content/chr19_ldl_pheno
MODEL_DIR=/content/chr19_ldl_pheno_result/models
SCORE_DIR=/content/chr19_ldl_pheno_result/scores

POWER=-0.25
HER=0.1
NUM_CAUSALS=512
PHENO_INDEX=2

for SPLIT_TEST in 'train' 'test'

do

    BED=${GENO_DIR}/${SUPERPOP}_${SPLIT_TEST}
    PHENO=${PHEN_DIR}/power=${POWER}_her=${HER}_num-causals=${NUM_CAUSALS}.pheno
    MODEL=${MODEL_DIR}/BayesR_${SUPERPOP}_train_PHENO${PHENO_INDEX}_power=${POWER}_her=${HER}_num-causals=${NUM_CAUSALS}.snpRes
    OUT=${SCORE_DIR}/BayesR_${SUPERPOP}_${SPLIT_TEST}_PHENO${PHENO_INDEX}_power=${POWER}_her=${HER}_num-causals=${NUM_CAUSALS}

    ./plink \
        --bfile $BED \
        --score $MODEL 2 5 8 header sum center \
        --out $OUT

done

"""
with open('script.sh', 'w') as file:
  file.write(sh)

! bash script.sh
```
