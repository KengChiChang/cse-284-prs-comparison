```python
import os

# Define the directory containing the files
# Assuming current directory, but you can modify this path as needed
directory = "."

# Function to swap the first two columns in a line
def swap_columns(line):
    columns = line.strip().split()  # Split by any whitespace
    # Swap the first two columns
    columns[0], columns[1] = columns[1], columns[0]
    return ' '.join(columns)

# Function to extract parameters from filename and construct new filename
def construct_new_filename(filename):
    if "HumDef" in filename:
        power = "-0.25"
    elif "GCTA" in filename:
        power = "-1"
    else:
        return None  # Return None if filename does not match expected format

    params_part = filename.split("params=")[1].split("}.pheno")[0]
    new_filename = f"power={power}_{params_part.replace(',', '_')}"
    return new_filename

# Iterate over all files in the specified directory
for filename in os.listdir(directory):
    print("Processing: ", filename)
    # Check if the file has the correct format and needs processing
    if filename.endswith(".pheno") and ("HumDef_params" in filename or "GCTA_params" in filename):
        filepath = os.path.join(directory, filename)

        # Read the content and swap columns
        with open(filepath, 'r') as file:
            lines = file.readlines()
            swapped_lines = [swap_columns(line) + '\n' for line in lines]

        # Write the content back after swapping columns
        with open(filepath, 'w') as file:
            file.writelines(swapped_lines)

        # Construct the new filename
        new_filename = construct_new_filename(filename)
        if new_filename:
            new_filepath = os.path.join(directory, new_filename + ".pheno")
            # Rename the file
            os.rename(filepath, new_filepath)

print("Files have been processed and renamed.")
```

```python
import os

def remove_curly_braces(filename):
    new_filename = filename.replace('{', '')  # Remove curly braces
    os.rename(filename, new_filename)

def rename_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".pheno"):
            new_filename = filename.replace('.pheno', '.phen')
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))

# Get the current directory
current_directory = os.getcwd()

# Remove curly braces from filenames and rename .pheno files to .phen
for filename in os.listdir(current_directory):
    if filename.endswith(".pheno"):
        filepath = os.path.abspath(os.path.join(current_directory, filename))
        remove_curly_braces(filepath)

# Rename .pheno files to .phen
rename_files(current_directory)


```
