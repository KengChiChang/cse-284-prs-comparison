```python
import pandas as pd
effects_df = pd.read_csv('ldl_prs.txt', sep='\t', names=['snp', 'allele', 'effect'])
effects_df = effects_df.drop(columns=['allele'])
effects_df
```

```python
chr19_snps = pd.read_csv('chr19.bim', sep="\t", names=["chr", "snp", "pos", "coord", "alt", "ref"])
chr19_snps
```

```python
effects_chr19 = effects_df.merge(chr19_snps, on='snp')
effects_chr19
```

```python
effects_chr19_transposed = effects_chr19.transpose()
effects_chr19_transposed
```

```python
pheno_name = "PHENOTEST"
causals = (effects_chr19_transposed
           .loc[["snp"]]
           .reset_index(drop=True))
causals
```

```python
effects = (effects_chr19_transposed
           .loc[["effect"]]
           .reset_index(drop=True))
effects
```

```python
! mkdir causals_effects
```

```python
repeat_size = 100
causals_replicate = causals.loc[causals.index.repeat(repeat_size)].reset_index(drop=True)
effects_replicate = effects.loc[effects.index.repeat(repeat_size)].reset_index(drop=True)
causals_replicate.to_csv('causals_effects/causals_all.txt', sep="\t", header=False, index=False)
effects_replicate.to_csv('causals_effects/effects_all.txt', sep="\t", header=False, index=False)
```

```python
sh = """

BFILE=chr19
CAUSALFILE=causals_effects/causals_all.txt
EFFECTSFILE=causals_effects/effects_all.txt
NUM_PHENO=100
NUM_CAUSL=512
PARENT_DIR="/content/chr19_ldl_pheno"


for POWER in {-1,-0.25}

do
    for HER in {0.1,0.3,0.5,0.7,0.9}

    do

    OUT=${PARENT_DIR}/power=${POWER}_her=${HER}_num-causals=${NUM_CAUSL}

        ./ldak5.2.linux \
            --make-phenos $OUT \
            --bfile $BFILE \
            --num-phenos $NUM_PHENO \
            --causals $CAUSALFILE \
            --effects $EFFECTSFILE \
            --power $POWER \
            --her $HER \
            --num-causals $NUM_CAUSL \
            --ignore-weights YES

    done

done

"""
with open('script.sh', 'w') as file:
  file.write(sh)

! bash script.sh
```

```python
num_causals = 10
effects_chr19.sample(n=num_causals, random_state=1)
```

```python
sample_size = 100

for num_causals in [1, 10, 50, 100, 250]:

    sample_causals_list = list()
    sample_effects_list = list()

    for i in range(sample_size):
        sampled_snps = (effects_chr19[['snp', 'effect']]
                        .sample(n=num_causals, random_state=i)
                        .transpose())
        sampled_causal_row = (sampled_snps
                            .loc[["snp"]]
                            .reset_index(drop=True)
                            .set_axis(range(sampled_snps.shape[1]), axis=1))
        sampled_effect_row = (sampled_snps
                            .loc[["effect"]]
                            .reset_index(drop=True)
                            .set_axis(range(sampled_snps.shape[1]), axis=1))
        sample_causals_list.append(sampled_causal_row)
        sample_effects_list.append(sampled_effect_row)

    sample_causals_df = pd.concat(sample_causals_list, axis=0, ignore_index=True)
    sample_effects_df = pd.concat(sample_effects_list, axis=0, ignore_index=True)

    sample_causals_df.to_csv('causals_effects/causals_num-causals='+str(num_causals)+'.txt',
                             sep="\t", header=False, index=False)
    sample_effects_df.to_csv('causals_effects/effects_num-causals='+str(num_causals)+'.txt',
                             sep="\t", header=False, index=False)
```

```python
sh = """

BFILE=chr19
NUM_PHENO=100
PARENT_DIR="/content/chr19_ldl_pheno"

for NUM_CAUSL in {1,10,50,100,250}

do

    CAUSALFILE=causals_effects/causals_num-causals=${NUM_CAUSL}.txt
    EFFECTSFILE=causals_effects/effects_num-causals=${NUM_CAUSL}.txt

    for POWER in {-1,-0.25}

    do
        for HER in {0.1,0.3,0.5,0.7,0.9}

        do

        OUT=${PARENT_DIR}/power=${POWER}_her=${HER}_num-causals=${NUM_CAUSL}

            ./ldak5.2.linux \
                --make-phenos $OUT \
                --bfile $BFILE \
                --num-phenos $NUM_PHENO \
                --causals $CAUSALFILE \
                --effects $EFFECTSFILE \
                --power $POWER \
                --her $HER \
                --num-causals $NUM_CAUSL \
                --ignore-weights YES

        done

    done

done

"""
with open('script.sh', 'w') as file:
  file.write(sh)

! bash script.sh
```
