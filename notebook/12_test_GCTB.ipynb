```python
import pandas as pd
import scipy.stats
print("CEU - test on GBR")

prs = pd.read_csv("CEU_BayesR_GBR.profile", delim_whitespace=True)
phen = pd.read_csv("/content/drive/MyDrive/CSE-284-Final-Project/data/test_data/kgvcf_ldl.phen",
                   delim_whitespace=True, names=["FID","IID","phen"])
d = pd.merge(prs, phen, on=["IID"])
print("R2=%s"%(scipy.stats.pearsonr(d["phen"],d["SCORESUM"])[0]**2))
```
